```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import L1L2
from keras import backend
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state = 0)
```

```python
model = Sequential()
input_num = X.shape[1]
```

```python
model.add(Dense(1, input_dim = input_num, activation = 'sigmoid', kernel_regularizer = L1L2(l2 = 0.3)))
```

```python
# https://keras.io/ko/getting-started/sequential-model-guide/

model.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(X_train, y_train, epochs = 500)
```

```python
# evaluate result : [loss, accuracy]
train_result = model.evaluate(X_train, y_train)
test_result = model.evaluate(X_test, y_test)
```

```python
regularizer = []
train_loss = []
train_accuracy = []
test_loss = []
test_accuracy = []

for i in range(10):
    l2 = i * 0.01
    model = Sequential()
    input_num = X.shape[1]
    model.add(Dense(1, input_dim = input_num, activation = 'sigmoid', kernel_regularizer = L1L2(l2 = l2)))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(X_train, y_train, epochs = 500, verbose = 0)
    
    train_result = model.evaluate(X_train, y_train, verbose = 0)
    test_result = model.evaluate(X_test, y_test, verbose = 0)
    
    regularizer.append(l2)
    train_loss.append(train_result[0])
    train_accuracy.append(train_result[1])
    test_loss.append(test_result[0])
    test_accuracy.append(test_result[1])
```

```python
plt.plot(regularizer, train_accuracy)
plt.plot(regularizer, test_accuracy)
plt.xlabel('regularizer')
plt.ylabel('accuracy')
plt.legend(['train', 'test'])
```

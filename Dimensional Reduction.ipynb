```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mglearn
%matplotlib inline
```

# Data Normalization

1) Standard scaler : mean = 0, variance = 1

2) Robust scaler : median = 0, use quartile value(Q1 = 25%, Q3 = 75%) / (X-median) / (X_Q3 - X_Q1) => less sensitive to outlier

3) MinMax scaler : 0 < x1 < 1, 0 < x2 < 1, min = 0, max = 0

4) Normalizer : make Eulidean distance 1, different angle

```python
mglearn.plots.plot_scaling()
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
X = cancer.data
y = cancer.target
```

```python
# MinMax scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(X)
X_scaled = scaler.transform(X)
```

```python
plt.scatter(X[:,0], X[:,1])
```

```python
plt.scatter(X_scaled[:,0], X_scaled[:,1])
```

```python
print('min : ', X_scaled.max(axis = 0)) # axis = 0 : feature
print('max : ', X_scaled.min(axis = 0)) # axis = 1 : sample
```

```python
# Standard scaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X) # StandardScaler 계산에 필요한 정보 학습(mean, variance)
X_scaled = scaler.transform(X) # StandardScaler에 맞추어 X domain transform
```

```python
print('mean : ', X_scaled.mean(axis = 0))
print('var : ', X_scaled.var(axis = 0))
```

```python
pd.DataFrame(X)
```

```python
pd.DataFrame(X_scaled)
```

# PCA(Principle Components Analysis)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10).fit(X_scaled)
X_scaled_pca = pca.transform(X_scaled)
print('before pca : ', X.shape)
print('after pca : ', X_scaled_pca.shape)
```

```python
mglearn.discrete_scatter(X_scaled_pca[:,0], X_scaled_pca[:,1], y)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
plt.plot(pca.explained_variance_)
```

```python
plt.plot(pca.explained_variance_ratio_)
```

# Factor Analysis

```python
from sklearn.decomposition import FactorAnalysis
fa = FactorAnalysis(n_components=10, random_state=0)
X_scaled_fa = fa.fit_transform(X_scaled)
print('X_scaled_fa : ', X_scaled_fa.shape)
```

```python
plt.scatter(X_scaled_fa[:,0], X_scaled_fa[:,1])
```
